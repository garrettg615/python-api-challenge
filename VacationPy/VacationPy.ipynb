```python
## dependecies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import gmaps
import os
from pprint import pprint

# Import API Key
from vaca_config import go_key
```

```python
## import csv file from Output folder and display
city_weather_df = pd.read_csv('../Output/weather_data.csv')
city_weather_df.head()
```

```python
## initialize and configure gmaps with api key
gmaps.configure(api_key=go_key)
```

```python
## set variables for Heat map
locations = city_weather_df[['Lat','Long']].astype(float)
city_humidity = city_weather_df['Humidity']

## initialize google map figure for heat map
fig = gmaps.figure()

## define heat map layer using humidity
heat_humidity = gmaps.heatmap_layer(locations, weights=city_humidity,
                                    max_intensity=50)

## Add heat may layer
fig.add_layer(heat_humidity)

## Show figure
fig
```

```python
## Define the parameters for the ideal vacation spots
ideal_locations_df = city_weather_df.loc[(city_weather_df['Max Temp']>=70)&
                                         (city_weather_df['Max Temp']<=80)&
                                         (city_weather_df['Wind Speed']<=10)&
                                         (city_weather_df['Cloudiness']==0)]

## add a blank column for hotel names for ideal location spots
ideal_locations_df['Hotel'] = ''
                

```

```python
## define base url for google places api
base_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?locations='

## define the paramaters for the google places api request
params = {'location':'',
         'radius':5000,
         'type':'hotel',
         'key':go_key}

## for loop through city name to get nearest hotel and name to display in dataframe
for index, row in ideal_locations_df.iterrows():
    params['location'] = str(row['Lat']) + ',' + str(row['Long'])
    response = requests.get(base_url, params=params).json()
    
    ideal_locations_df.loc[index, 'Hotel'] = response['results'][1]['name']
    
```

```python
## display data frame for ideal vacation spots with first hotel name
ideal_locations_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in ideal_locations_df.iterrows()]
locations = ideal_locations_df[["Lat", "Long"]]
```

```python
## create markers with hotel name, city and country
markers = gmaps.marker_layer(locations,info_box_content=hotel_info)
fig.add_layer(markers)
fig
```
